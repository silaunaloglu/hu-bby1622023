```python
# Katalogdaki eserleri listeleme fonksiyonu
def listele():
    with open("veri.txt", "r") as f:
        for line in f:
            print(line)
            

# Kataloga yeni bir eser ekleme fonksiyonu
def ekle():
    adi = input("Eser adı: ")
    yazar = input("Yazar adı: ")
    basim_yili = input("Basım yılı: ")
    with open("veri.txt", "a") as f:
        f.write(f"{adi}, {yazar}, {basim_yili}\n")
        print('Eser başarılı bir şekilde klendi')

# Katalog içerisinde seçilen eseri silme fonksiyonu
def sil():
    eser = input("Silinecek eserin adını giriniz: ")
    with open("veri.txt", "r") as f:
        lines = f.readlines()
    with open("veri.txt", "w") as f:
        for line in lines:
            if not eser in line:
                f.write(line)

# Katalog içerisinde seçilen eseri güncelleme fonksiyonı
def guncelle():
    eser = input("Güncellenecek eserin adını giriniz: ")
    yeni_adi = input("Yeni eser adı: ")
    yeni_yazar = input("Yeni yazar adı: ")
    yeni_basim_yili = input("Yeni basım yılı: ")
    with open("veri.txt", "r") as f:
        lines = f.readlines()
    with open("veri.txt", "w") as f:
        for line in lines:
            if eser in line:
                f.write(f"{yeni_adi}, {yeni_yazar}, {yeni_basim_yili}\n")

# Katalogu tamamen temizleme fonksiyonu
def temizle():
    with open("veri.txt", "w") as f:
        f.write("")
        print('Katalog temizlendi.')

# Katalog içerisinde eser aratma fonkisyonu.
def arama():
    kelime = input("Aranacak kelimeyi giriniz: ")
    with open("veri.txt", "r") as f:
        for line in f:
            if kelime in line:
                print(line)        
            


def menu():
    while True:
        print("1 - Eserleri listele")
        print("2 - Yeni eser ekle")
        print("3 - Eser sil")
        print("4 - Eser güncelle")
        print("5 - Katalogu temizle")
        print("6 - Arama yap")
        print("0 - Çıkış yap")
        
        secim = input("Yapmak istediğiniz işlemi seçiniz: ")
    
        if secim == "1":
            listele()
        elif secim == "2":
            ekle()
        elif secim == "3":
            sil()
        elif secim == "4":
            guncelle()
        elif secim == "5":
            temizle()
        elif secim == "6":
            arama()
        elif secim == "0":
            kaydet()
            break
        else:
            print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin.")

def kaydet():
    print("Değişiklikler kaydediliyor...")

menu()
```
